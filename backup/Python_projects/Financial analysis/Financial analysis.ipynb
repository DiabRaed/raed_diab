```python
#this creates a new column with the close price of tomorrow
fb['price1']=fb['Close'].shift(-1)
fb
```

```python
#We wanna look at price difference between close of today and close of tomorrow (which is what we wanna predict)
fb['price difference']=fb['price1']-fb['Close']
fb
```

```python
#or we can calculate the daily return 
fb['daily return']=fb['price difference']/fb['Close']
fb
```

```python
#let us look at the direction: +1 donates going up, -1 donates going down
fb['direction']=[1 if fb.loc[ei,'price difference']>0 else -1 for ei in fb.index]
fb
```

```python
#let's calculate the moving average of every 3 days 
fb['moving average']=(fb['Close']+fb['Close'].shift(1)+fb['Close'].shift(2))/3
fb
```

```python
#dataframe has rolling() method to calculate the moving average nicely
fb['moving10']=fb['Close'].rolling(10).mean()
fb['moving50']=fb['Close'].rolling(50).mean()

fb
```

```python
#moving20 is a fast signal because it's closer to the real price
plt.figure(figsize=(10, 8))
fb['Close'].plot()
fb['moving10'].plot()
fb['moving50'].plot()
```

```python
#we wanna see if should be a share or not based on the moving average for the fewer number of days 
#this is 1 when we should buy a stock and 0 when we shouldn't
fb['shares']=[1 if fb.loc[ei,'moving10']>fb.loc[ei,'moving50'] else 0 for ei in fb.index]
fb
```

```python
#create daily profit which is the close price of tomorrow - price of today.. this works only when shares=1
#that is we have a stock in hand 
plt.figure(figsize=(10, 8))

fb['close1']=fb['Close'].shift(-1)
fb['profit']=[fb.loc[ei,'close1']-fb.loc[ei,'Close'] if fb.loc[ei,'shares']==1 else 0 for ei in fb.index]
fb['profit'].plot()
plt.axhline(y=0,color="red")
```

```python
#or we can compute the cumulative wealth
plt.figure(figsize=(10, 8))

fb['wealth']=fb['profit'].cumsum()
fb
# fb.tail()
# fb['wealth'].plot()

```

```python
plt.figure(figsize=(10, 8))

print("Total money I spent " ,fb.loc[fb.index[0],'Close'])
print("Total money I won ", fb.loc[fb.index[-2],'wealth'])
fb['wealth'].plot()
```

```python
apple = pd.read_csv('/Users/Financial analysis/AAPL.csv',index_col=0)
```

```python
apple.loc['2018-01-01':'2018-12-31','Close'].plot()
```

```python
apple['logreturn']=np.log(apple['Close'].shift(-1))-np.log(apple['Close'])
apple['logreturn'].hist(bins=50)
# apple.loc['2018-01-01':'2019-01-01','logreturn'].plot()
```

```python
from scipy.stats import norm
mu = apple['logreturn'].mean()
sigma = apple['logreturn'].std(ddof=1)

density = pd.DataFrame()
density['x'] = np.arange(apple['logreturn'].min()-0.01, apple['logreturn'].max()+0.01, 0.001)
density['pdf'] = norm.pdf(density['x'], mu, sigma)

apple['logreturn'].hist(bins=50, figsize=(15, 8))
plt.plot(density['x'], density['pdf'], color='red')
plt.show()
```

```python
# the probablity of a return of -5% is 
prob_return1 = norm.cdf(-0.05, mu, sigma)
print('The Probability is ', prob_return1)
```

```python
# drop over 40% in 220 days
mu220 = 220*mu
sigma220 = (220**0.5) * sigma
print('The probability of dropping over 40% in 220 days is ', norm.cdf(-0.4, mu220, sigma220))
```

```python
# Lets build 90% confidence interval for log return
sample_size = apple['logreturn'].shape[0]
sample_mean = apple['logreturn'].mean()
sample_std = apple['logreturn'].std(ddof=1) / sample_size**0.5

# left and right quantile
z_left = norm.ppf(0.05)
z_right = norm.ppf(0.95)

# upper and lower bound
interval_left = sample_mean+z_left*sample_std
interval_right = sample_mean+z_right*sample_std
```

```python
# 90% confidence interval tells you that there will be 90% chance that the average stock return lies between "interval_left"
# and "interval_right".

print('90% confidence interval is ', (interval_left, interval_right))
```

```python
#let's calculate z-hat and t-hat distribution curves
xbar=apple['logreturn'].mean()
s=apple['logreturn'].std(ddof=1)
n=apple['logreturn'].shape[0]
mu=0  #the null assertion is the average return is zero which is mu=0
```

```python
zhat=(xbar-mu)/(s/(n**0.5))     #if z>>0 then we can reject the null and the average daily return isn't zero
zhat
```

```python
#to set the criterion at which we assume zhat>>0
# confidence level
alpha = 0.05

zleft = norm.ppf(alpha/2, 0, 1)
zright = -zleft  # z-distribution is symmetric 
print(zleft, zright)
```

```python
print('At significant level of {}, shall we reject: {}'.format(alpha, zhat>zright or zhat<zleft))
```
